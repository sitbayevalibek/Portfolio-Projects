# This small program helps to get the products from the technomart.uz online store in the form of a CSV file.

```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = 'https://texnomart.uz/uz/katalog/smartfony'
page_count = 17  # Sizning bergan URL lar soni

file_name = 'telefonlar.csv'

# Mahsulotlar ma'lumotlarini CSV faylga yozish
with open(file_name, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Mahsulot nomi', 'Narxi', 'Rasm manzili', 'Brend', 'Model', 'Og\'irligi',
                     "Korpus turi",
                     "SIM-kartalar soni",
                     "Bir necha SIM-kartalarni ishlash rejimi",
                     "Video yozib olish",
                     "ОS turi",
                     "Protsessor",
                     'Interfeyslar',
                     "Diagonal",
                     "Rang",
                     "O'lchamlari (ExВxS) mm",
                     "Chaqmoq",
                     "Ekran turi",
                     "SIM-karta turi",
                     "Quvvatlanish uyasini turi",
                     "Tomonlar nisbati",
                     "Sensorli ekranni turi",
                     "1 dyuymga piksellar soni (PPI)",
                     "Ekranni avtomatik burish",
                     "Ichki xotira (ROM)",
                     "Yadrolar soni",
                     "Tezkor xotira (RAM)",
                     "Fotokamera",
                     "Old kamera",
                     "Akkumulyator hajmi"
                     ])

    for page in range(1, page_count + 1):
        url = base_url + f'?page={page}'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        products = soup.find_all('div', class_='product-bottom')

        for product in products:
            product_url = product.find('a', class_='product-name')['href']
            product_response = requests.get('https://texnomart.uz' + product_url)
            product_soup = BeautifulSoup(product_response.text, 'html.parser')

            name = product_soup.find('h1', class_='product__name').text.strip()
            price = product_soup.find('div', class_='product__price').text.strip()
            image_url = product_soup.find('img')['src']

            characteristics = product_soup.find('div', id='product-char-wrap').find_all('div', class_='list__item')

            brend = ''
            model = ''
            ogirligi = ''
            Korpus_turi = ''
            SIM_kartalar_soni = ''
            Bir_necha_SIM_kartalarni_ishlash_rejimi = ''
            Video_yozib_olish = ''
            os_turi = ''
            Protsessor = ''
            Interfeyslar = ''
            Diagonal = ''
            Rang = ''
            Olchamlari_mm = ''
            Chaqmoq = ''
            Ekran_turi = ''
            SIM_karta_turi = ''
            Quvvatlanish_uyasini_turi = ''
            Tomonlar_nisbati = ''
            Sensorli_ekranni_turi = ''
            dyuymga_piksellar_soni = ''
            Ekranni_avtomatik_burish = ''
            Ichki_xotira = ''
            Yadrolar_soni = ''
            Tezkor_xotira = ''
            Fotokamera = ''
            Old_kamera = ''
            Akkumulyator_hajmi = ''

            for item in characteristics:
                item_name = item.find('div', class_='list__name').text.strip()
                item_value = item.find('div', class_='list__value').text.strip()
                if item_name == 'Brend':
                    brend = item_value
                elif item_name == 'Model':
                    model = item_value
                elif item_name == "Og'irligi (gr)":
                    ogirligi = item_value
                elif item_name == "Korpus turi":
                    Korpus_turi = item_value
                elif item_name == "SIM-kartalar soni":
                    SIM_kartalar_soni = item_value
                elif item_name == "Bir necha SIM-kartalarni ishlash rejimi":
                    Bir_necha_SIM_kartalarni_ishlash_rejimi = item_value
                elif item_name == "Video yozib olish":
                    Video_yozib_olish = item_value
                elif item_name == "ОS turi":
                    os_turi = item_value
                elif item_name == "Protsessor":
                    Protsessor = item_value
                elif item_name == 'Interfeyslar':
                    Interfeyslar = item_value
                elif item_name == "Diagonal":
                    Diagonal = item_value
                elif item_name == "Rang":
                    Rang = item_value
                elif item_name == "O'lchamlari (ExВxS) mm":
                    Olchamlari_mm = item_value
                elif item_name == "Chaqmoq":
                    Chaqmoq = item_value
                elif item_name == "Ekran turi":
                    Ekran_turi = item_value
                elif item_name == "SIM-karta turi":
                    SIM_karta_turi = item_value
                elif item_name == "Quvvatlanish uyasini turi":
                    Quvvatlanish_uyasini_turi = item_value
                elif item_name == "Tomonlar nisbati":
                    Tomonlar_nisbati = item_value
                elif item_name == "Sensorli ekranni turi":
                    Sensorli_ekranni_turi = item_value
                elif item_name == "1 dyuymga piksellar soni (PPI)":
                    dyuymga_piksellar_soni = item_value
                elif item_name == "Ekranni avtomatik burish":
                    Ekranni_avtomatik_burish = item_value
                elif item_name == "Ichki xotira (ROM)":
                    Ichki_xotira = item_value
                elif item_name == "Yadrolar soni":
                    Yadrolar_soni = item_value
                elif item_name == "Tezkor xotira (RAM)":
                    Tezkor_xotira = item_value
                elif item_name == "Fotokamera":
                    Fotokamera = item_value
                elif item_name == "Old kamera":
                    Old_kamera = item_value
                elif item_name == "Akkumulyator hajmi":
                    Akkumulyator_hajmi = item_value

            writer.writerow([name, price, image_url, brend, model, ogirligi, Korpus_turi, SIM_kartalar_soni,
                             Bir_necha_SIM_kartalarni_ishlash_rejimi, Video_yozib_olish, os_turi, Protsessor,
                             Interfeyslar, Diagonal, Rang, Olchamlari_mm, Chaqmoq, Ekran_turi, SIM_karta_turi,
                             Quvvatlanish_uyasini_turi, Tomonlar_nisbati, Sensorli_ekranni_turi, dyuymga_piksellar_soni,
                             Ekranni_avtomatik_burish, Ichki_xotira, Yadrolar_soni, Tezkor_xotira, Fotokamera,
                             Old_kamera, Akkumulyator_hajmi])

print(f'Mahsulotlar CSV fayliga saqlandi: {file_name}')
```
